–ú–æ–ª–æ–¥–µ—Ü, —á—Ç–æ –∏—Å–ø–æ–ª—å–∑—É–µ—à—å –º–µ—Ç–æ–¥ dropna, –∞ —Ç–∞–∫–∂–µ –ø–∏—à–µ—à—å, –∫–∞–∫—É—é —á–∞—Å—Ç—å –¥–∞—Ç–∞—Å–µ—Ç–∞ —Ç—ã —É–¥–∞–ª—è–µ—à—å ‚Äì —ç—Ç–æ –ø–æ–º–æ–≥–∞–µ—Ç –Ω–µ –ø–µ—Ä–µ—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ
</div>

–û–ø—Ä–µ–¥–µ–ª—è—é –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤

```python

def categorize_locality(row):
    try:
        if '–ø–æ—Å–µ–ª–æ–∫' in row:
            return '–ø–æ—Å–µ–ª–æ–∫'
        elif '—Å–µ–ª–æ' in row:
            return '—Å–µ–ª–æ'
        elif '–¥–µ—Ä–µ–≤–Ω—è' in row:
            return '–¥–µ—Ä–µ–≤–Ω—è'
        elif '—Ç–æ–≤–∞—Ä–∏—â–µ—Å—Ç–≤–æ' in row:
            return '—Ç–æ–≤–∞—Ä–∏—â–µ—Å—Ç–≤–æ'
        elif '–ü–∞–Ω—Å–∏–æ–Ω–∞—Ç' in row:
            return '–ø–∞–Ω—Å–∏–æ–Ω–∞—Ç'
        else:
            return '–≥–æ—Ä–æ–¥'
    except:
        return '–≥–æ—Ä–æ–¥'
    
data = data.apply(lambda x: x.replace({'—ë':'–µ'}, regex=True))
data['locality_type'] = data['locality_name'].apply(categorize_locality)

#data.info()
```

<div class="alert alert-info"> <b>–ú–æ–∏ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏: </b> –ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ —Å —ç—Ç–∞–∂–Ω–æ—Å—Ç—å—é –¥–æ–º–∞, –µ—Å–ª–∏ –∏–∑–≤–µ—Å—Ç–Ω–æ –Ω–∞ –∫–∞–∫–æ–º —ç—Ç–∞–∂–µ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∞ –∫–≤–∞—Ä—Ç–∏—Ä–∞. –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞—é –≤–∑—è—Ç—å 7 –Ω–∞–∏–±–æ–ª–µ–µ —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω—ë–Ω–Ω—ã—Ö —ç—Ç–∞–∂–Ω–æ—Å—Ç–µ–π –∏ –æ–∫—Ä—É–≥–ª—è—Ç—å –¥–æ –±–ª–∏–∂–∞–π—à–µ–≥–æ.</div>

```python
def categorize_floor(floor):
    try:
        if floor == 1:
            return 1
        elif 1 <= floor <= 5:
            return 5
        elif 6 <= floor <= 12:
            return 12
        elif 13 <= floor <= 17:
            return 17
        elif 18 <= floor <= 25:
            return 25
        elif 26 <= floor <= 33:
            return 33
        elif 34 <= floor <= 50:
            return 50
        elif floor > 50:
            return 100
    except:
        pass

rows = data['floors_total'].isna()
data.loc[rows, 'floors_total'] = data.loc[rows, 'floor'].apply(categorize_floor)
data[rows]
```

–ó–∞–º–µ–Ω–∞ —Ç–∏–ø–∞ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –¥–∞—Ç—ã –∏–∑ —Å—Ç—Ä–æ–∫–æ–≤–æ–π –≤ datetime –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞ –æ–ø–µ—Ä–∞—Ü–∏–π –≤ –¥–∞–ª—å–Ω–µ–π—à–µ–º

–î–æ–±–∞–≤–ª—é –ø–æ–ª—è "–¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏" - weekday, "–º–µ—Ü—è—Ü" - month –∏ "–≥–æ–¥" - year, –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è


```python
data['first_day_exposition'] = pd.to_datetime(
    data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')

data['weekday'] = data['first_day_exposition'].dt.weekday 

data['month'] = pd.DatetimeIndex(data['first_day_exposition']).month

data['year'] = pd.DatetimeIndex(data['first_day_exposition']).year

data.sample(3)
```

```python
# –ò–∑–º–µ–Ω—è—é —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –¥–ª—è days_exposition
data.days_exposition = data.days_exposition.astype('int')

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Å–¥–≤–∏–≥–∞ –¥–∞—Ç—ã
def last_day(item):
    first_day_exposition = item['first_day_exposition']
    days_exposition = item['days_exposition']
    return first_day_exposition + pd.DateOffset(days=days_exposition)

# –°–æ–∑–¥–∞—é —Å—Ç–æ–ª–±–µ—Ü —Å –¥–∞—Ç–æ–π —Å–Ω—è—Ç–∏—è –æ–±—ä—è–≤–ª–µ–Ω–∏—è
data['last_day_exposition'] = data.apply(last_day, axis=1)

# –°–æ–∑–¥–∞—é —Å—Ç–æ–ª–±–µ—Ü —Å –¥–Ω—ë–º –Ω–µ–¥–µ–ª–∏ —Å–Ω—è—Ç–∏—è –æ–±—ä—è–≤–ª–µ–Ω–∏—è
data['last_weekday'] = data['last_day_exposition'].dt.weekday

# –ü—Ä–∏–º–µ—Ä –æ–±–Ω–æ–≤–ª—ë–Ω–Ω–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö 
data.sample(3)

#data.info()
```

–ö–∞—Ç–µ–≥–æ—Ä–∏—è —ç—Ç–∞–∂–∞:

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ —ç—Ç–∞–∂–∞
def floors_group(item):
    #display(item)
    floors_total = item['floors_total']
    floor = item['floor']
    if floors_total == floor:
        return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    elif floor == 1:
        return '–ø–µ—Ä–≤—ã–π'
    return '–¥—Ä—É–≥–æ–π'
```

```python
# –î–æ–±–∞–≤–ª–µ–Ω–∏—è –ø–æ–ª—è —Å –∫–∞—Ç–µ–≥–æ—Ä–∏–µ–π —ç—Ç–∞–∂–∞
data['floors_group'] = data.apply(floors_group, axis=1)
# –ü—Ä–∏–º–µ—Ä –æ–±–Ω–æ–≤–ª—ë–Ω–Ω–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö 
data.sample(3)
```
